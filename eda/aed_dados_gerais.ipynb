### Análise dos valores dos subsídios repassados às famílias acolhedoras

```python
df_dados_gerais['q14'].apply(type).unique()
```

```python
df_dados_gerais['q14'].isna().sum()
```

```python
df_dados_gerais['q14'].fillna(0, inplace=True)
```

```python
df_dados_gerais['q14'] = pd.to_numeric(df_dados_gerais['q14'], errors='coerce')
```

```python
# Depois de analisar o primeiro gráfico, nota-se que o ES apresentou outliers significativos
df_dados_gerais[df_dados_gerais['q0_10'] == 'ES']['q14'].describe()
```

```python
# Calcula o valor do 75º percentil para 'q14' na UF 'ES'
percentil_75_es = df_dados_gerais[df_dados_gerais['q0_10'] == 'ES']['q14'].quantile(0.75)

# Exclui valores acima do 75º percentil para 'ES'
df_sem_outliers = df_dados_gerais[~((df_dados_gerais['q0_10'] == 'ES') & (df_dados_gerais['q14'] > percentil_75_es))]

# Filtra o DataFrame original para listar as linhas com valores de 'q14' acima do 75º percentil
linhas_acima_75_es = df_dados_gerais[(df_dados_gerais['q0_10'] == 'ES') & (df_dados_gerais['q14'] > percentil_75_es)]

# Exibe o DataFrame com os outliers
linhas_acima_75_es
```

```python
# Exclui valores acima do 75º percentil para 'ES'
percentil_75_es = df_dados_gerais[df_dados_gerais['q0_10'] == 'ES']['q14'].quantile(0.75)
df_sem_outliers = df_dados_gerais[~((df_dados_gerais['q0_10'] == 'ES') & (df_dados_gerais['q14'] > percentil_75_es))]
```

```python
# Definindo um dicionário que mapeia cada UF para sua região
ufs_para_regiao = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AP': 'Norte', 'AM': 'Norte', 'BA': 'Nordeste',
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}

# Cria uma nova coluna 'Região' no DataFrame usando o mapeamento
df_sem_outliers['Região'] = df_sem_outliers['q0_10'].map(ufs_para_regiao)
```

```python
# Calcula a média por UF
media_por_uf_sem_outliers = df_sem_outliers.groupby(['q0_10', 'Região'])['q14'].mean().reset_index()
media_por_uf_sem_outliers
```

```python
# Define cores para cada região
cores_regiao = {
    'Norte': 'blue', 'Nordeste': 'green', 'Centro-Oeste': 'orange',
    'Sudeste': 'red', 'Sul': 'purple'
}
```

```python
# Cria o gráfico de barras
plt.figure(figsize=(12, 8))
for regiao, cor in cores_regiao.items():
    subset = media_por_uf_sem_outliers[media_por_uf_sem_outliers['Região'] == regiao]
    plt.bar(subset['q0_10'], subset['q14'], label=regiao, color=cor)

# Configurações do gráfico
plt.title('Média mensal de subsídio repassado às famílias acolhedoras')
plt.xlabel('UF')
plt.ylabel('Média mensal de subsídio')
plt.xticks(rotation=45)
plt.legend(title='Região')
plt.tight_layout()
plt.show()
```

```python
media_por_uf_sem_outliers.describe()
```

#### Há correlação entre valor médio de subsídio repassado e a quantidade de famílias aptas/que estejam acolhendo?

```python
#q46 é o número de famílias acolhedoras cadastradas no serviço (que estejam aptas a acolher e/ou estejam acolhendo)
df_dados_gerais['q46'].apply(type).unique()
```

```python
df_dados_gerais['q46'].isna().sum()
```

```python
df_dados_gerais['q46'].describe()
```

```python
df_dados_gerais['Região'] = df_dados_gerais['q0_10'].map(ufs_para_regiao)
```

```python
# Calcula a média de famílias aptas/acolhendo por UF
media_qtd_fam = df_dados_gerais.groupby(['q0_10', 'Região'])['q46'].mean().reset_index()
```

```python
# Cria o gráfico de barras para média de famílias aptas/acolhendo
plt.figure(figsize=(12, 8))
for regiao, cor in cores_regiao.items():
    subset = media_qtd_fam[media_qtd_fam['Região'] == regiao]
    plt.bar(subset['q0_10'], subset['q46'], label=regiao, color=cor)

# Configurações do gráfico
plt.title('Quantidade média de famílias aptas/que estão acolhendo por UF')
plt.xlabel('UF')
plt.ylabel('Quantidade média de famílias')
plt.xticks(rotation=45)
plt.legend(title='Região')
plt.tight_layout()
plt.show()
```

```python
subsidio_qtd_fam = pd.merge(media_por_uf_sem_outliers, media_qtd_fam, on=['q0_10', 'Região'])
```

```python
corr_subsidio_qtd_fam = subsidio_qtd_fam[['q14', 'q46']].corr().iloc[0, 1]
print("Coeficiente de Correlação entre Subsídio e Quantidade de Famílias:", corr_subsidio_qtd_fam)
```

```python
# Scatter plot para visualizar a correlação
plt.figure(figsize=(8, 6))
plt.scatter(subsidio_qtd_fam['q46'], subsidio_qtd_fam['q14'], alpha=0.7)
plt.title('Correlação entre Subsídio e Quantidade média de famílias aptas')
plt.xlabel('Quantidade média de famílias')
plt.ylabel('Subsídio Médio (Reais)')
plt.grid(True)
plt.show()
```
